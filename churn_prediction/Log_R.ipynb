```python
var_list = ['PhoneService','PaperlessBilling','Churn','Partner','Dependents']
```

```python
def binary_map(x):
    return x.map({'Yes' : 1, 'No' : 0})

df[var_list] = df[var_list].apply(binary_map)
```
